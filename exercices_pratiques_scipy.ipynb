### Énoncé
Vous travaillez sur un projet de simulation physique. Créez un programme qui calcule :

1. L'énergie d'un photon de lumière bleue (λ = 450 nm) en joules et en électron-volts
   - Formule : E = h × c / λ
   - h : constante de Planck, c : vitesse de la lumière

2. La force gravitationnelle entre la Terre et la Lune
   - Formule : F = G × (m₁ × m₂) / r²
   - Masse Terre : 5.972e24 kg, Masse Lune : 7.342e22 kg
   - Distance : 384,400 km

3. Convertir une pression de 2.5 atmosphères en :
   - Pascals
   - PSI (pounds per square inch)
   - Bars

### Indices
- Utilisez `scipy.constants` pour toutes les constantes

````python
from scipy import constants
#qst 1
lamda = 450
lamda_metre = lamda * 1e-9
h = constants.h
c = constants.c
electron_volt = constants.eV

E = h * c / lamda_metre
E_ev = E / electron_volt

print(f"L'énergie d'un photon de lumière bleue : {lamda} nm ")
print(f"E = {E:.3e} J")
print(f"E = {E_ev:.3e} eV")
````

````python
#qst3
P_atm = 2.5
P_pascals = P_atm * constants.atm
P_psi = P_pascals / constants.psi
P_bar = P_pascals / constants.bar

print(f"P = {P_atm:.3f} atmosphères")
print(f"P = {P_pascals:.3e} Pa")
print(f"P = {P_psi:.3f} psi")
print(f"P = {P_bar:.3f} bar")
````

## Exercice 1.3 : Optimisation d'un modèle de vente
**Niveau : ⭐⭐ Intermédiaire**

### Énoncé
Une entreprise marocaine de production d'huile d'argan souhaite optimiser son prix de vente.

**Données collectées** (prix en DH, quantité en litres) :
```
Prix :     [150, 180, 200, 220, 250, 280, 300]
Quantité : [850, 780, 720, 650, 580, 500, 420]
```

**Tâches :**

1. Utilisez `curve_fit` pour ajuster un modèle polynomial de degré 2 : `Q(p) = a×p² + b×p + c`

2. Calculez le revenu R(p) = p × Q(p) et trouvez le prix optimal qui maximise le revenu
   - Utilisez `minimize` avec la fonction négative du revenu

3. Visualisez sur un graphique :
   - Les points de données réels
   - La courbe ajustée Q(p)
   - Le point de revenu maximal

4. Affichez le prix optimal et le revenu maximal attendu

### Contraintes
- Le prix doit être entre 100 DH et 400 DH
- Utilisez `scipy.optimize.minimize` avec méthode='L-BFGS-B'


````python
import numpy as np
from scipy import optimize
prix = np.array([150, 180, 200, 220, 250, 280, 300])
quantite = np.array([850, 780, 720, 650, 580, 500, 420])

#qst1
def modele_polynomial(p, a, b, c):
  return a * p**2 + b * p + c

params, covariance = optimize.curve_fit(modele_polynomial, prix, quantite)
a, b, c = params

print(f"Modèle ajusté : Q(p) = {a:.6f}×p² + {b:.4f}×p + {c:.2f}")
````

````python
#qst 2

def Revenu(p):
  Q = modele_polynomial(p,a,b,c)
  return p * Q

result = optimize.minimize(lambda p: -Revenu(p), x0=200, method='L-BFGS-B' ,bounds=[(100, 400)])
prix_optimal = result.x[0]
revenu_max = -result.fun

print(f"\nPrix optimal: {prix_optimal:.2f} DH")
print(f"Revenu maximal attendu: {revenu_max:.2f} DH")
print(f"Quantité à ce prix: {modele_polynomial(prix_optimal, a, b, c):.0f} litres")
````

````python
#qst 3
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(prix, quantite, color='red', s=100, label='Données réelles')
plt.plot(prix, modele_polynomial(prix, a, b, c), color='blue', label='Modèle ajusté')

plt.xlabel('Prix')
plt.ylabel('Quantité')
plt.title('Données réelles et modèle ajusté')
plt.legend()

````

````python
plt.subplot(1, 2, 2)
prix_continu = np.linspace(100, 400, 100)
quantite_ajustee = modele_polynomial(prix_continu, a, b, c)

revenu_continu = prix_continu *   quantite_ajustee

plt.plot(prix_continu, revenu_continu, 'purple', linewidth=2, label='Revenu R(p)')
plt.scatter([prix_optimal], [revenu_max], color='green', s=200, marker='*',
           label=f'Maximum ({prix_optimal:.0f} DH, {revenu_max:.0f} DH)')
plt.xlabel('Prix (DH)', fontsize=11)
plt.ylabel('Revenu (DH)', fontsize=11)
plt.title('Courbe de revenu', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
````

## Exercice 2.1 : Analyse statistique d'un dataset médical
**Niveau : ⭐⭐ Intermédiaire**

### Énoncé
Vous analysez les données de tension artérielle de 30 patients avant et après un traitement.

**Données :**
```python
avant = [145, 138, 152, 148, 155, 142, 149, 150, 147, 151,
         143, 146, 154, 141, 148, 153, 144, 147, 150, 146,
         149, 152, 145, 148, 151, 143, 147, 150, 144, 149]

apres = [138, 132, 145, 140, 148, 135, 142, 143, 140, 144,
         136, 139, 147, 134, 141, 146, 137, 140, 143, 139,
         142, 145, 138, 141, 144, 136, 140, 143, 137, 142]
```

**Tâches :**

1. Calculez pour chaque groupe (avant/après) :
   - Moyenne, médiane, écart-type
   - Utiliser `scipy.stats.describe()`

2. Testez si la différence est statistiquement significative :
   - Test t de Student pour échantillons appariés (`ttest_rel`)
   - Interprétez la p-value (seuil α = 0.05)

3. Vérifiez la normalité des distributions :
   - Utilisez `shapiro` test
   - Tracez les histogrammes

4. Calculez la corrélation entre avant et après (`pearsonr`)

5. Rédigez une conclusion : le traitement est-il efficace ?


````python
from scipy import stats
import numpy as np

#qst 1
avant = np.array([145, 138, 152, 148, 155, 142, 149, 150, 147, 151,
                  143, 146, 154, 141, 148, 153, 144, 147, 150, 146,
                  149, 152, 145, 148, 151, 143, 147, 150, 144, 149])

apres = np.array([138, 132, 145, 140, 148, 135, 142, 143, 140, 144,
                  136, 139, 147, 134, 141, 146, 137, 140, 143, 139,
                  142, 145, 138, 141, 144, 136, 140, 143, 137, 142])

desc_avant = stats.describe(avant)
desc_apres = stats.describe(apres)
print("Statistique descriptive avant")
print(f"  Moyenne: {desc_avant.mean:.2f} mmHg")
print(f"  Médiane: {np.median(avant):.2f} mmHg")
print(f"  Écart-type: {np.sqrt(desc_avant.variance):.2f}")

print("Statistique descriptive apres ")
print(f"  Moyenne: {desc_apres.mean:.2f} mmHg")
print(f"  Médiane: {np.median(apres):.2f} mmHg")
print(f"  Écart-type: {np.sqrt(desc_apres.variance):.2f}")
````

````python
#qst 2
t_statistic, p_value = stats.ttest_rel(avant, apres)
print(f"  Statistique t: {t_statistic:.4f}")
print(f"  P-value: {p_value:.6f}")
print(f"  Seuil α: 0.05")
if p_value < 0.05:
    print(f"  Différence significative (p < 0.05)")
else:
    print(f"  Différence NON significative (p ≥ 0.05)")
````

````python
#qst 3
shapiro_avant = stats.shapiro(avant)
shapiro_apres = stats.shapiro(apres)

print(f"shapiro test:")
print(f"  Avant: p-value = {shapiro_avant.pvalue:.4f} {'(normal)' if shapiro_avant.pvalue > 0.05 else '(non-normal)'}")
print(f"  Après: p-value = {shapiro_apres.pvalue:.4f} {'(normal)' if shapiro_apres.pvalue > 0.05 else '(non-normal)'}")
````

````python
#qst 4
correlation, p_corr = stats.pearsonr(avant, apres)
print(f"  Coefficient: {correlation:.4f}")
print(f"  P-value: {p_corr:.6f}")
````

## Exercice 2.4 : Matrices creuses et graphes

### Énoncé

Vous analysez un réseau de transport entre 6 villes.

**Tâches :**

1. **Créez la matrice d'adjacence** du réseau (distances en km) :
   ```
   Villes : A(0), B(1), C(2), D(3), E(4), F(5)
   
   Liaisons :
   A-B: 50, A-C: 80
   B-C: 30, B-D: 100
   C-D: 70, C-E: 90
   D-E: 40, D-F: 60
   E-F: 50
   ```
   (Le graphe est non-orienté : A-B = B-A)

2. **Convertissez** la matrice en format CSR

3. **Analyse de connectivité :**
   - Vérifiez que le graphe est connexe avec `connected_components()`

4. **Plus court chemin :**
   - Utilisez `dijkstra()` pour trouver le plus court chemin de A vers F
   - Affichez la distance totale et le chemin emprunté

5. **Visualisation :**
   - Affichez la matrice des distances sous forme de heatmap

````python
#qst 1
n_villes = 6
distances = np.zeros((n_villes, n_villes))

liaisons = [
    (0, 1, 50), (0, 2, 80),  # A-B, A-C
    (1, 2, 30), (1, 3, 100),  # B-C, B-D
    (2, 3, 70), (2, 4, 90),   # C-D, C-E
    (3, 4, 40), (3, 5, 60),   # D-E, D-F
    (4, 5, 50)                # E-F
]

for i, j, dist in liaisons:
    distances[i, j] = dist
    distances[j, i] = dist  # Symétrie

print("1. Matrice d'adjacence (distances en km):")
villes = ['A', 'B', 'C', 'D', 'E', 'F']
print("   ", "  ".join(villes))
for i, ville in enumerate(villes):
    print(f"{ville}: {distances[i]}")

#qst 2
matrice_csr = sparse.csr_matrix(distances)
print(f"\n2. Matrice convertie en format CSR:")
print(f"  Éléments non-nuls: {matrice_csr.nnz}")

#qst 3
n_comp, labels = connected_components(matrice_csr, directed=False)
print(f"\n3. Analyse de connectivité:")
print(f"  Nombre de composantes: {n_comp}")
print(f"  Le graphe est {'CONNEXE' if n_comp == 1 else 'NON CONNEXE'} ✓")

# qst 4
ville_depart = 0
ville_arrivee = 5
dist_matrix, predecessors = dijkstra(matrice_csr, indices=ville_depart,
                                     directed=False, return_predecessors=True)
distance_totale = dist_matrix[ville_arrivee]

chemin = []
current = ville_arrivee
while current != -9999:
    chemin.append(current)
    current = predecessors[current]
    if current == ville_depart:
        chemin.append(current)
        break
chemin = chemin[::-1]
chemin_noms = [villes[i] for i in chemin]

print(f"\n4. Plus court chemin de {villes[ville_depart]} vers {villes[ville_arrivee]}:")
print(f"  Chemin: {' → '.join(chemin_noms)}")
print(f"  Distance totale: {distance_totale:.0f} km")

````

````python
tree = spatial.KDTree(clients)
print(f"\n2. Analyse avec KDTree:")

for i in range(min(3, len(clients))):
    # 3 voisins les plus proches (en excluant le point lui-même)
    distances, indices = tree.query(clients[i], k=4)
    print(f"\n   Client {i}:")
    print(f"   - 3 voisins plus proches: {indices[1:]} à distances {distances[1:]}")

    # Voisins dans un rayon de 0.02 degrés
    indices_rayon = tree.query_ball_point(clients[i], r=0.02)
    print(f"   - Clients dans rayon 0.02°: {len(indices_rayon)-1} clients")
````

signal_bruite = voix + bruit_hf + bruit_bf + bruit_random

**Tâches :**

1. Supprimez la tendance du signal avec `signal.detrend`

   - Identifiez les pics de fréquence

3. Filtrez le signal :
   - Supprimez les fréquences < 80 Hz (bruit électrique)

4. Reconstruisez le signal avec la FFT inverse

5. Visualisez (4 sous-graphiques) :
   - Signal original bruité
   - Spectre avant filtrage
   - Spectre après filtrage
   - Signal nettoyé

6. Calculez le rapport signal/bruit (SNR) avant et après


````python
# Signal original : voix humaine (fréquences 100-300 Hz)
t = np.linspace(0, 10, 10000)
voix = np.sin(2 * np.pi * 150 * t) + 0.5 * np.sin(2 * np.pi * 250 * t)

# Bruit haute fréquence (ventilation, équipement)
bruit_hf = 0.8 * np.sin(2 * np.pi * 2000 * t)

# Bruit basse fréquence (ronronnement électrique 50 Hz)
bruit_bf = 0.6 * np.sin(2 * np.pi * 50 * t)

# Bruit aléatoire
bruit_random = np.random.normal(0, 0.3, len(t))

# Signal enregistré
signal_bruite = voix + bruit_hf + bruit_bf + bruit_random
````

````python
from scipy import signal
#qst1
signal_detrend = signal.detrend(signal_bruite)
print(f"Signal nettoyé: {signal_detrend.shape}")
````

````python
#qst 2
fft_bruite = np.fft.fft(signal_detrend)
freqs = np.fft.fftfreq(len(t), t[1] - t[0])
puissance_bruite = np.abs(fft_bruite) ** 2

indices_positifs = freqs > 0
freqs_pos = freqs[indices_positifs]
puissance_pos = puissance_bruite[indices_positifs]
top_freqs_idx = np.argsort(puissance_pos)[-5:]
print(f"   Top 5 fréquences dominantes: {freqs_pos[top_freqs_idx]} Hz")
````

````python
#qst 3
fft_filtre = fft_bruite.copy()
masque_freq = (np.abs(freqs) < 80) | (np.abs(freqs) > 400)
fft_filtre[masque_freq] = 0
````

````python
#qst 4
signal_nettoye = np.fft.ifft(fft_filtre).real
puissance_filtre = np.abs(fft_filtre) ** 2
print(f"   Signal nettoyé: {signal_nettoye.shape}")
````

````python
#qst 6
def calculer_snr(signal, bruit):
    puissance_signal = np.mean(signal ** 2)
    puissance_bruit = np.mean(bruit ** 2)
    return 10 * np.log10(puissance_signal / puissance_bruit)

bruit_total = signal_bruite - voix
bruit_residuel = signal_nettoye - voix

snr_avant = calculer_snr(voix, bruit_total)
snr_apres = calculer_snr(voix, bruit_residuel)

print(f"\nRapport Signal/Bruit (SNR):")
print(f"   Avant filtrage: {snr_avant:.2f} dB")
print(f"   Après filtrage: {snr_apres:.2f} dB")
print(f"   Amélioration: {snr_apres - snr_avant:.2f} dB ")
````

## Exercice 3.2 : Analyse d'un signal avec la FFT
**Niveau : ⭐⭐ Intermédiaire**

### Énoncé

Vous analysez un signal audio composé de plusieurs fréquences mélangées avec du bruit.

**Tâches :**

1. **Création du signal :**
   ```python
   np.random.seed(42)
   fs = 1000  # Fréquence d'échantillonnage (Hz)
   t = np.linspace(0, 1, fs)  # 1 seconde
   
   # Signal composé de 3 sinusoïdes : 50 Hz, 120 Hz, 300 Hz
   signal_pur = 3*np.sin(2*np.pi*50*t) + 1.5*np.sin(2*np.pi*120*t) + 0.8*np.sin(2*np.pi*300*t)
   
   # Ajout de bruit
   bruit = np.random.randn(len(t)) * 2
   signal_bruite = signal_pur + bruit
   ```

2. **Analyse spectrale :**
   - Calculez la FFT du signal bruité
   - Calculez les fréquences associées avec `fftfreq()`
   - Identifiez les fréquences dominantes (pics dans le spectre)

3. **Filtrage :**
   - Mettez à zéro les composantes de faible amplitude (< seuil) dans le spectre
   - Reconstruisez le signal filtré avec `ifft()`

4. **Visualisation (3 subplots) :**
   - Signal original bruité
   - Spectre de fréquences (amplitude vs fréquence, uniquement les fréquences positives)
   - Signal après filtrage

````python
np.random.seed(42)
fs = 1000  # Fréquence d'échantillonnage (Hz)
t = np.linspace(0, 1, fs)  # 1 seconde

# Signal composé de 3 sinusoïdes : 50 Hz, 120 Hz, 300 Hz
signal_pur = 3*np.sin(2*np.pi*50*t) + 1.5*np.sin(2*np.pi*120*t) + 0.8*np.sin(2*np.pi*300*t)

# Ajout de bruit
bruit = np.random.randn(len(t)) * 2
signal_bruite = signal_pur + bruit
````

````python
#qst 2
fft_result = np.fft.fft(signal_bruite)
freqs_fft = np.fft.fftfreq(len(t), 1/fs)
amplitude_spectrum = np.abs(fft_result)

positive_freqs = freqs_fft[:len(freqs_fft)//2]
positive_amplitudes = amplitude_spectrum[:len(amplitude_spectrum)//2]

threshold = np.max(positive_amplitudes) * 0.1
peaks_indices = np.where(positive_amplitudes > threshold)[0]
dominant_freqs = positive_freqs[peaks_indices]
print(f"   Fréquences dominantes détectées: {dominant_freqs[:5]} Hz")
````

````python
#qst 3
seuil_amplitude = np.max(amplitude_spectrum) * 0.05
fft_filtered = fft_result.copy()
fft_filtered[amplitude_spectrum < seuil_amplitude] = 0

signal_filtre = np.fft.ifft(fft_filtered).real

print(f"   Seuil: {seuil_amplitude:.2f}")
print(f"   Composantes conservées: {np.sum(amplitude_spectrum >= seuil_amplitude)}")
````

````python
#qst 4
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Signal original bruité
axes[0].plot(t, signal_bruite, 'b-', alpha=0.7, linewidth=0.8)
axes[0].set_xlabel('Temps (s)', fontsize=10)
axes[0].set_ylabel('Amplitude', fontsize=10)
axes[0].set_title('Signal original bruité', fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Spectre de fréquences
axes[1].plot(positive_freqs, positive_amplitudes, 'r-', linewidth=1)
axes[1].axhline(seuil_amplitude, color='green', linestyle='--',
                label=f'Seuil de filtrage')
axes[1].set_xlabel('Fréquence (Hz)', fontsize=10)
axes[1].set_ylabel('Amplitude', fontsize=10)
axes[1].set_title('Spectre de fréquences', fontweight='bold')
axes[1].set_xlim(0, 500)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Signal après filtrage
axes[2].plot(t, signal_filtre, 'g-', alpha=0.9, linewidth=0.8)
axes[2].set_xlabel('Temps (s)', fontsize=10)
axes[2].set_ylabel('Amplitude', fontsize=10)
axes[2].set_title('Signal après filtrage', fontweight='bold')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
````

````python
#qst 1
image_size = 256
image_radio = np.ones((image_size, image_size)) * 100  # Fond gris

# Ajout de cercles (poumons + anomalie)
y, x = np.ogrid[:image_size, :image_size]
# Cercle 1: poumon gauche
mask1 = (x - 80)**2 + (y - 128)**2 <= 50**2
image_radio[mask1] = 150
# Cercle 2: poumon droit
mask2 = (x - 176)**2 + (y - 128)**2 <= 50**2
image_radio[mask2] = 150
# Cercle 3: anomalie
mask3 = (x - 128)**2 + (y - 80)**2 <= 15**2
image_radio[mask3] = 200

# Ajout de bruit
bruit_image = np.random.normal(0, 10, image_radio.shape)
image_radio_bruitee = image_radio + bruit_image

````

````python
#qst 5
structure_elem = np.ones((3, 3))
contours_erodes = ndimage.binary_erosion(contours_fermes, structure=structure_elem)
contours_dilates = ndimage.binary_dilation(contours_fermes, structure=structure_elem)
````

````python
#Partie 1
from sklearn.datasets import load_iris
iris_data = load_iris()
X_iris = iris_data.data
y_iris = iris_data.target
feature_names = iris_data.feature_names
target_names = iris_data.target_names

print(f"\nDataset Iris chargé:")
print(f"  Échantillons: {X_iris.shape[0]}")
print(f"  Features: {X_iris.shape[1]}")
print(f"  Classes: {len(target_names)} - {list(target_names)}")
````

````python
print("\n1. Statistiques descriptives par feature:")
for i, name in enumerate(feature_names):
    desc = stats.describe(X_iris[:, i])
    print(f"\n{name}:")
    print(f"  Moyenne: {desc.mean:.2f}")
    print(f"  Écart-type: {np.sqrt(desc.variance):.2f}")
    print(f"  Min-Max: ({desc.minmax[0]:.2f}, {desc.minmax[1]:.2f})")
````

````python
# 2. Test de normalité
print("\n2. Tests de normalité (Shapiro-Wilk):")
for i, name in enumerate(feature_names):
    stat, p = stats.shapiro(X_iris[:, i])
    print(f"  {name}: p={p:.4f} {'(normal)' if p > 0.05 else '(non-normal)'}")

# 3. Matrice de corrélation
corr_matrix = np.corrcoef(X_iris.T)
print("\n3. Matrice de corrélation:")
print("   ", " ".join([f"{fn[:8]:>8s}" for fn in feature_names]))
for i, name in enumerate(feature_names):
    print(f"{name[:8]:8s}", " ".join([f"{corr_matrix[i, j]:8.3f}" for j in range(4)]))
````

````python
#Partie 2
# Split train/test (80/20)
np.random.seed(42)
indices = np.random.permutation(len(X_iris))
split_point = int(0.8 * len(X_iris))
train_indices = indices[:split_point]
test_indices = indices[split_point:]

X_train = X_iris[train_indices]
y_train = y_iris[train_indices]
X_test = X_iris[test_indices]
y_test = y_iris[test_indices]

print(f"\nSplit des données:")
print(f"  Train: {len(X_train)} échantillons")
print(f"  Test: {len(X_test)} échantillons")

# Fonction KNN avec KDTree
def knn_predict(X_train, y_train, X_test, k=5):
    tree = spatial.KDTree(X_train)
    predictions = []

    for x_test in X_test:
        distances, indices = tree.query(x_test, k=k)
        # Ensure indices is always a 1D array to correctly index y_train
        # If k=1, indices is an int. np.atleast_1d makes it an array like [index].
        # If k>1, indices is already an array.
        voisins_labels = y_train[np.atleast_1d(indices)]
        # Vote majoritaire
        prediction = np.bincount(voisins_labels).argmax()
        predictions.append(prediction)

    return np.array(predictions)

# Test avec k=5
k_initial = 5
y_pred = knn_predict(X_train, y_train, X_test, k=k_initial)
accuracy_initial = np.mean(y_pred == y_test)

print(f"\n Classification avec k={k_initial}:")
print(f"   Accuracy: {accuracy_initial*100:.2f}%")
````

````python
#PARTIE 3
k_values = range(1, 21)
accuracies = []

for k in k_values:
    y_pred_k = knn_predict(X_train, y_train, X_test, k=k)
    acc = np.mean(y_pred_k == y_test)
    accuracies.append(acc)

best_k = k_values[np.argmax(accuracies)]
best_accuracy = max(accuracies)

print(f"\nOptimisation terminée:")
print(f"  Meilleur k: {best_k}")
print(f"  Meilleure accuracy: {best_accuracy*100:.2f}%")
````

````python
# Partie 4
# Prédictions finales avec best_k
y_pred_final = knn_predict(X_train, y_train, X_test, k=best_k)

# 1. Matrice de confusion
conf_matrix = np.zeros((3, 3), dtype=int)
for true, pred in zip(y_test, y_pred_final):
    conf_matrix[true, pred] += 1

print(f"\n1. Matrice de confusion (k={best_k}):")
print("   Prédictions →")
print("Réalité ↓  ", " ".join([f"{tn:>10s}" for tn in target_names]))
for i, name in enumerate(target_names):
    print(f"{name:10s}", " ".join([f"{conf_matrix[i, j]:10d}" for j in range(3)]))

# 2. Métriques par classe
print("\n2. Métriques par classe:")
for i, class_name in enumerate(target_names):
    TP = conf_matrix[i, i]
    FP = conf_matrix[:, i].sum() - TP
    FN = conf_matrix[i, :].sum() - TP
    TN = conf_matrix.sum() - TP - FP - FN

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    rappel = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * rappel) / (precision + rappel) if (precision + rappel) > 0 else 0

    print(f"\n   {class_name}:")
    print(f"     Précision: {precision:.3f}")
    print(f"     Rappel: {rappel:.3f}")
    print(f"     F1-score: {f1:.3f}")

# 3. Test statistique Chi2 pour comparer les performances
chi2_stat = 0
for i in range(3):
    observed = conf_matrix[i, i]
    expected = len(y_test) / 3
    chi2_stat += (observed - expected)**2 / expected

print(f"\n3. Test Chi2 pour équilibre des performances:")
print(f"   Statistique: {chi2_stat:.3f}")
````

Vous travaillez sur un projet de maintenance prédictive. Vous disposez de données de vibration d'une machine industrielle. Vous devez analyser ces signaux pour détecter des anomalies.

**Données simulées :**
```python
np.random.seed(42)

# Paramètres
fs = 1000  # Fréquence d'échantillonnage
duration = 2  # 2 secondes
t = np.linspace(0, duration, fs * duration)

# Signal normal : vibration à 100 Hz avec légères harmoniques
signal_normal = (2 * np.sin(2*np.pi*100*t) +
                 0.5 * np.sin(2*np.pi*200*t) +

# Signal anormal : apparition d'une fréquence parasite à 250 Hz (défaut)
signal_anormal = (2 * np.sin(2*np.pi*100*t) +
                  0.5 * np.sin(2*np.pi*200*t) +
                  1.5 * np.sin(2*np.pi*250*t) +  # Fréquence de défaut
                  np.random.randn(len(t)) * 0.3)
```

**Tâches :**

1. **Prétraitement :**
   - Supprimez la tendance des deux signaux avec `signal.detrend()`

2. **Analyse spectrale comparative :**
   - Calculez la FFT des deux signaux
   - Identifiez les fréquences dominantes de chaque signal

3. **Détection d'anomalie :**
   - Définissez un seuil pour détecter les pics significatifs dans le spectre
   - Identifiez la fréquence anormale (250 Hz) qui n'existe que dans le signal défaillant

4. **Statistiques descriptives :**
   - Utilisez `stats.describe()` pour comparer les deux signaux
   - Comparez les variances et les valeurs extrêmes

5. **Filtrage du signal anormal :**
   - Supprimez la composante à 250 Hz du spectre
   - Reconstruisez le signal "réparé"

6. **Visualisation complète (3 lignes × 2 colonnes) :**
   - Ligne 1 : Signaux temporels (normal vs anormal)
   - Ligne 2 : Spectres de fréquence (normal vs anormal)
   - Ligne 3 : Signal anormal original vs signal filtré

````python
#QST1
np.random.seed(42)
fs_iot = 1000
duration_iot = 2
t_iot = np.linspace(0, duration_iot, fs_iot * duration_iot)

# Signal normal
signal_normal = (2 * np.sin(2*np.pi*100*t_iot) +
                 0.5 * np.sin(2*np.pi*200*t_iot) +
                 np.random.randn(len(t_iot)) * 0.3)

# Signal anormal
signal_anormal = (2 * np.sin(2*np.pi*100*t_iot) +
                  0.5 * np.sin(2*np.pi*200*t_iot) +
                  1.5 * np.sin(2*np.pi*250*t_iot) +  # Fréquence de défaut
                  np.random.randn(len(t_iot)) * 0.3)

print("\nSignaux de vibration machine générés:")
print(f"  Signal normal: vibration à 100 Hz + harmoniques (200 Hz)")
print(f"  Signal anormal: + fréquence parasite à 250 Hz (défaut)")

# === 1. PRÉTRAITEMENT ===
print("\n1. Prétraitement - Suppression de tendance")
signal_normal_detrend = signal.detrend(signal_normal)
signal_anormal_detrend = signal.detrend(signal_anormal)
print("   ✓ Tendances supprimées")
````

````python
#qst 2
# FFT des deux signaux
fft_normal = np.fft.fft(signal_normal_detrend)
fft_anormal = np.fft.fft(signal_anormal_detrend)
freqs_iot = np.fft.fftfreq(len(t_iot), 1/fs_iot)

# Spectres de puissance
puissance_normal = np.abs(fft_normal) ** 2
puissance_anormal = np.abs(fft_anormal) ** 2

# Fréquences positives seulement
idx_pos = freqs_iot > 0
freqs_pos_iot = freqs_iot[idx_pos]
puiss_norm_pos = puissance_normal[idx_pos]
puiss_anom_pos = puissance_anormal[idx_pos]

# Identifier fréquences dominantes
seuil_detection = np.max(puiss_norm_pos) * 0.05
pics_normal = freqs_pos_iot[puiss_norm_pos > seuil_detection]
pics_anormal = freqs_pos_iot[puiss_anom_pos > seuil_detection]

print(f"   Fréquences dominantes (normal): {pics_normal[:5]} Hz")
print(f"   Fréquences dominantes (anormal): {pics_anormal[:5]} Hz")

````

````python
#qst 3
# Différence de spectre
diff_spectre = puiss_anom_pos - puiss_norm_pos
anomalies_idx = np.where(diff_spectre > np.max(diff_spectre) * 0.3)[0]
freq_anomalies = freqs_pos_iot[anomalies_idx]

print(f"   Fréquences anormales détectées: {freq_anomalies[:5]} Hz")
print(f"   ✓ Fréquence de défaut identifiée: ~250 Hz")
````

````python
#qst 4
desc_normal = stats.describe(signal_normal_detrend)
desc_anormal = stats.describe(signal_anormal_detrend)

print(f"   Signal normal:")
print(f"     Moyenne: {desc_normal.mean:.4f}")
print(f"     Variance: {desc_normal.variance:.4f}")
print(f"     Min-Max: ({desc_normal.minmax[0]:.2f}, {desc_normal.minmax[1]:.2f})")

print(f"\n   Signal anormal:")
print(f"     Moyenne: {desc_anormal.mean:.4f}")
print(f"     Variance: {desc_anormal.variance:.4f}")
print(f"     Min-Max: ({desc_anormal.minmax[0]:.2f}, {desc_anormal.minmax[1]:.2f})")
````

````python
#qst5
#Supprimer la composante à 250 Hz
fft_repare = fft_anormal.copy()
# Masque pour supprimer 250 Hz ± 10 Hz
masque_250Hz = (np.abs(freqs_iot) >= 240) & (np.abs(freqs_iot) <= 260)
fft_repare[masque_250Hz] = 0

# Reconstruction
signal_repare = np.fft.ifft(fft_repare).real
puissance_repare = np.abs(fft_repare) ** 2

print(f"   ✓ Composante 250 Hz supprimée")
print(f"   Variance avant: {np.var(signal_anormal_detrend):.4f}")
print(f"   Variance après: {np.var(signal_repare):.4f}")
````
